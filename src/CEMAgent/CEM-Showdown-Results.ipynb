# Visualizing DQN RL Agent Results in Pokemon Showdown

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
```

## Training Process

```python
random_train = pd.read_csv('RandomPlayer_trainlog_20000eps.csv')
random_train.drop('Unnamed: 0', axis=1, inplace=True)
random_train_stats = random_train.describe().round(2)
random_train_stats
```

```python
plt.figure(figsize=(16,4), dpi=100)
plt.scatter(random_train['nb_steps'], random_train['episode_reward'], alpha = 0.7)
plt.title("CEM Agent Trained Against RandomPlayer", size=15)
plt.xlabel('Number of Steps')
plt.ylabel('Episode Reward')
plt.show()

plt.hist(random_train['episode_reward'], alpha = 0.7)
plt.title(f"Reward Histogram of CEM Agent Trained Against Random Player \n Mean reward: {random_train_stats['episode_reward']['mean']}", size=15)
plt.xlabel('Episode Reward')
plt.ylabel('Count')
```

## Testing Process

```python
random_test = pd.read_csv('(RandomPlayer_20000)RandomPlayer_testlog_100eps.csv')
random_test.drop('Unnamed: 0', axis=1, inplace=True)
random_stats = random_test.describe().round(2)
random_stats
```

```python
max_test = pd.read_csv('(RandomPlayer_20000)MaxPlayer_testlog_100eps.csv')
max_test.drop('Unnamed: 0', axis=1, inplace=True)
max_stats = max_test.describe().round(2)
max_stats
```

```python
plt.figure(figsize=(16,4), dpi=100)

plt.subplot(1,2,1)
plt.hist(random_test['episode_reward'], alpha = 0.7)
plt.title(f"Eval: Reward of Trained CEM Agent vs RandomPlayer \n Mean reward: {random_stats['episode_reward']['mean']}", size=15)
plt.xlabel('Episode Reward')
plt.ylabel('Count')

plt.subplot(1,2,2)
plt.hist(max_test['episode_reward'], alpha = 0.7, color='coral')
plt.title(f"Eval: Reward of Trained CEM Agent vs MaxDamagePlayer \n Mean reward: {max_stats['episode_reward']['mean']}", size=15)
plt.xlabel('Episode Reward')
plt.ylabel('Count')
```
