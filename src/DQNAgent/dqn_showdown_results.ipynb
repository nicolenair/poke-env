# Visualizing DQN RL Agent Results in Pokemon Showdown

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
```

## Training Process

```python
# functions for creating the df and plots

def create_df(csv_file):
    '''
    create df from csv file
    '''
    train = pd.read_csv(csv_file)
    train.drop('Unnamed: 0', axis=1, inplace=True)
    return train

def create_plots(train_log, train_opponent, plot_color='blue'):
    '''
    Create train plots
    '''
    plt.figure(figsize=(16,4), dpi=120)
    plt.subplot(1,2,1)
    plt.scatter(train_log['nb_steps'], train_log['episode_reward'], color = plot_color, alpha = 0.7)
    plt.title(f"DQN Agent Trained Against {train_opponent}Player", size=15)
    plt.xlabel('Number of Steps')
    plt.ylabel('Episode Reward')

    plt.subplot(1,2,2)
    plt.hist(train_log['episode_reward'], color = plot_color, alpha = 0.7)
    plt.title(f"Reward Histogram of DQN Agent Trained Against {train_opponent} Player \n Mean reward: {round(train_log['episode_reward'].mean(), 2)}", size=15)
    plt.xlabel('Episode Reward')
    plt.ylabel('Count')
    plt.show()
```

```python
random_train = create_df('RandomPlayer_trainlog_20000eps.csv')
create_plots(random_train, 'Random', 'SteelBlue')
```

```python
max_train = create_df('MaxPlayer_trainlog_20000eps.csv')
create_plots(max_train, 'Max', 'coral')
```

```python
rl_train = create_df('PretrainedRLPlayer_trainlog_20000eps.csv')
create_plots(rl_train, 'PretrainedDQN', 'orange')
```

# Testing Process


## Trained Random Agent: Agent trained against Random Opponent 

### Trained Random Agent vs Random Opponent

The plots are shown in the following order:

- *Trained Random Agent vs Random Opponent*
- *Trained Random Agent vs Max Opponent*
- *Trained Random Agent vs Pretrained DQN Opponent*

```python
random_vs_random = create_df('TrainedRandomPlayer(20000)vsRandomPlayer_testlog_1000eps.csv')
create_plots(random_vs_random, 'Random vs Random', 'SteelBlue')

random_vs_max = create_df('TrainedRandomPlayer(20000)vsMaxPlayer_testlog_1000eps.csv')
create_plots(random_vs_max, 'Random vs Max', 'coral')

random_vs_rl = create_df('TrainedRandomPlayer(20000)vsRLPlayer_testlog_1000eps.csv')
create_plots(random_vs_rl, 'Random vs Pretrained DQN', 'orange')
```
